# *(Feature Engineering)*

**Будем обучать модель для предсказания популярности песни по её характеристикам, поэтому проведём соответствующую предобработку данных**

### Preprocessing:

```python
data.info()
```

```python
data = data.drop(["Unnamed: 0", "track_id", "track_name", "album_name"], axis=1)
```

```python
data.head()
```

```python
data["popularity"] = pd.qcut(data["popularity"], q=2, labels=[0, 1])
```

```python
data.head()
```

### One-Hot Encoding:

```python
def onehot_encode(df, column, prefix):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix=prefix)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    
    return df
```

```python
data = onehot_encode(data, "track_genre", "genre")
data = onehot_encode(data, "artists", 'artist')
```

```python
data
```

# *(Эксперименты с моделями машинного обучения/глубокого обучения*)

### Splitting and Scaling:

```python
y = data.loc[:, "popularity"][:1000]
X = data.drop("popularity", axis=1)[:1000]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=20)
```

### Modeling and Training:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
```

```python
log_model = LogisticRegression()
knn_model = KNeighborsClassifier()
dec_model = DecisionTreeClassifier()
mlp_model = MLPClassifier()
svm_model = SVC()
```

```python
log_model.fit(X_train, y_train)
knn_model.fit(X_train, y_train)
dec_model.fit(X_train, y_train)
mlp_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
```

```python
log_acc = log_model.score(X_test, y_test)
knn_acc = knn_model.score(X_test, y_test)
dec_acc = dec_model.score(X_test, y_test)
mlp_acc = mlp_model.score(X_test, y_test)
svm_acc = svm_model.score(X_test, y_test)
```

```python
print("Logistic Regression Accuracy:", log_acc)
print("K-Nearest-Neighbors Accuracy:", knn_acc)
print("Decision Tree Accuracy:", dec_acc)
print("Neural Network Accuracy:", mlp_acc)
print("Support Vector Machine Accuracy:", svm_acc)
```

```python
import plotly.express as px
```

```python
x=["Logistic Regression", "K-Nearest-Neighbors", "Decision Tree", "Neural Network", "Support Vector Machine"]
y=[log_acc, knn_acc, dec_acc, mlp_acc, svm_acc]

plt.figure(figsize=(10,6))
bars = plt.bar(x, y, color=['lightblue', 'lightyellow', 'lightgreen', 'tomato', 'pink'])

plt.xlabel("Модель", fontsize=12)
plt.ylabel("Точность", fontsize=12)
plt.title("Сравнение точности моделей", fontsize=14)

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=10)
   

plt.tight_layout()
plt.show()
```

**Обучив разные модели, лучшие результаты точности показали модель логистической регрессии и метода опорных векторов. Что говорит о том, что данные хорошо разделимы, имеют простую строктуру с линейными зависимостями и низким уровнем шума, что идеально подходит для этих методов.**
